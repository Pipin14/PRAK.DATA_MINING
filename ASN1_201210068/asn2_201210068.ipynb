## Import Library

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

## Load Dataset

```python
# Load the dataset
df = pd.read_csv('assessment 2 - StudentsPerformance.csv')
print(df.head())
```

## Clear the Dataset 

```python
# Add a delimiter=';' to the file to match the contents of the data to the column
df = pd.read_csv('assessment 2 - StudentsPerformance.csv', delimiter=';')

# Show data
df.head()
```

## Data visualization: Original data

```python
# Create bar plots for visualization
# Calculate the total based on the parents' education level
education_totals = df['parental level of education'].value_counts()

plt.figure(figsize=(7, 5))
education_totals.plot(kind='bar', color='skyblue')
plt.title('Total Berdasarkan Level Pendidikan Orang Tua')
plt.xlabel('Level Pendidikan Orang Tua')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create Pie chart diagram
# Calculate the total based on the parent's education level
education_totals = df['parental level of education'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(education_totals, labels=education_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Berdasarkan "Parental Level of Education"')
plt.axis('equal')
plt.show()
```

```python
# Sorts the DataFrame by the 'Total' column from highest to lowest value and displays the first few rows
df_sorted = df.assign(Total=df[['math score', 'reading score', 
                                'writing score']].sum(axis=1)).sort_values(by='Total', ascending=False)
df_sorted.head(10)
```

## Change nominal to numeric type file in the category column

```python
category_column = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

label_encoder = LabelEncoder()

for column in category_column:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

## Data visualization: Based on numeric data

```python
# Create bar plots for visualization
# Calculate the total based on the parents' education level
education_totals = df['parental level of education'].value_counts()

plt.figure(figsize=(7, 5))
education_totals.plot(kind='bar', color='skyblue')
plt.title('Total Berdasarkan Level Pendidikan Orang Tua')
plt.xlabel('Level Pendidikan Orang Tua')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create Pie chart diagram
# Calculate the total based on the parent's education level
education_totals = df['parental level of education'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(education_totals, labels=education_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Berdasarkan "Parental Level of Education"')
plt.axis('equal')
plt.show()
```

## Change decimal values to integers

```python
Column_update = ['math score', 'reading score', 'writing score']

for column in Column_update:
    df[column] = df[column].fillna(0).round().astype(int)

df.head()
```
